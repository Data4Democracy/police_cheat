```python
value_count_data = [(col, df[col].value_counts()) for col in df]# [0:250]
fig, ax = plt.subplots(len(value_count_data)//7 + 1,7, figsize=(20,len(value_count_data)*.5))
fig.tight_layout()
ax = ax.flatten()
for i, (col, answers) in enumerate(value_count_data):
    if len(answers)>10:
        ax[i].hist(answers, color='g', bins=40)
    else:
        answers.index = [10 if x==-9 else x for x in answers.index]
        ax[i].bar(answers.index, answers)
        ax[i].set_title(col)
```
